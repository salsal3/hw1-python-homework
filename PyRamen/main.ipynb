## Setup imports and csv

```python
# @TODO: Import libraries
import csv
from pathlib import Path
```

```python
# @TODO: Set file paths for menu_data.csv and sales_data.csv
menuPath = Path('../PyRamen/Resources/menu_data.csv')
salesPath = Path('../PyRamen/Resources/sales_data.csv')
```

## Read the data

### Menu data

```python
# @TODO: Initialize list objects to hold our menu and sales data
menu = []
sales = []

# @TODO: Read in the menu data into the menu list
# Use a `with` statement and open the `menu_data.csv` by using its file path
with open(menuPath, 'r') as csvfile:
    # Use the `reader` function from the `csv` library to begin reading `menu_data.csv`
    menuReader = csv.reader(csvfile, delimiter=',')
    
    # Use the `next` function to skip the header (first row of the CSV)
    menuHeader = next(menuReader)

    # Loop over the rest of the rows and append every row to the `menu` list object (the outcome will be a list of lists)
    for row in menuReader:
        menu.append(row)

menu[:3]
```

### Sales data

```python
# @TODO: Read in the sales data into the sales list
# Use a `with` statement and open the `sales_data.csv` by using its file path
with open(salesPath, 'r') as csvfile:
    # Use the `reader` function from the `csv` library to begin reading `menu_data.csv`
    salesReader = csv.reader(csvfile, delimiter=',')
    
    # Use the `next` function to skip the header (first row of the CSV)
    salesHeader = next(salesReader)
    print(salesHeader)

    # Loop over the rest of the rows and append every row to the `sales` list object (the outcome will be a list of lists)
    for row in salesReader:
        sales.append(row)

sales[:3]
```

## Manipulate the data

### Create a 'report' dictionary

```python
# @TODO: Initialize an empty 'report' dictionary to hold our key-value pairs of items and metrics
report = {}

# Initialize a row counter variable
row_count = 0
```

```python
# @TODO: Loop over every row in the sales list object
for row in sales:
    # @TODO: Initialize sales data variables
    quantity = int(row[3])
    sales_item = row[4]

    # If the item value is not in the report, add it as a new entry with initialized metrics
    if sales_item not in report.keys():
        report[sales_item] = {
                '01-count': 0,
                '02-revenue': 0,
                '03-cogs': 0,
                '04-profit': 0,
        }
        
    # For every row in sales data, loop over the menu records to determine a match
    for row in menu:
        
        # Initialize menu data variables
        menu_item = row[0]
        price = int(float(row[3]))
        cost = int(row[4])
        
        # Calculate profit of each item in the menu data
        profit = (price - cost)
        
        # @TODO: If the item value in the sales data is equal to any of the items in the menu, begin tracking metrics for that item
        if sales_item in row:
            
            # @TODO: Print out matching menu data
            # print(f'\n{row}\n')
            
            # @TODO: Cumulatively add up the metrics for each item key
            report[sales_item]['01-count'] += quantity
            report[sales_item]['02-revenue'] += price * quantity
            report[sales_item]['03-cogs'] += cost * quantity
            report[sales_item]['04-profit'] += profit * quantity
            
        # @TODO: Else, the sales item does not equal any of the items in the menu data, therefore no match
        else:
            continue
            # print(f'{sales_item} does not equal {menu_item}! NO MATCH!')
        
    # @TODO: Increment the row counter by 1
    row_count += 1
    print(f'Row Counter: {row_count}')
```

```python
# @TODO: Print total number of records in sales data
print(f'Total number of records: {len(sales)}')
```

```python
# @TODO: Write out report to a text file
with open('report.txt', 'w') as f:
    for key, value in report.items():
        f.write('%s:%s\n' % (key, value))
```
